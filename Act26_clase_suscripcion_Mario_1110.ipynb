```python
class Suscripcion:
    def __init__(self, usuario, plan, precio):
        # Inicialización de los 3 atributos solicitados
        self.usuario = usuario
        self.plan = plan
        self.precio = precio

    # Función 1: Lectura de datos mediante inputs
    def leerdatos(self):
        print(f"\n--- Registro de Suscripción ---")
        self.usuario = input("Ingrese el nombre del usuario: ")
        self.plan = input("Ingrese el tipo de plan: ")
        self.precio = input("Ingrese el precio: ")

    # Función 2: Muestra de datos procesados
    def mostrardatos(self):
        print("\n================================")
        print("       DETALLES DE CUENTA       ")
        print("================================")
        print(f"CLIENTE: {self.usuario}")
        print(f"PLAN:    {self.plan}")
        print(f"COSTO:   ${self.precio}")
        print("--------------------------------")

# --- EJECUCIÓN DEL PROGRAMA ---

# 1. Creamos los 2 objetos con valores vacíos o temporales
sub1 = Suscripcion("", "", "")
sub2 = Suscripcion("", "", "")

# 2. Llamada a funciones para el objeto sub1
sub1.leerdatos()
sub1.mostrardatos()

# 3. Llamada a funciones para el objeto sub2
sub2.leerdatos()
sub2.mostrardatos()

# Pie de página del programa
print("\nSistema finalizado exitosamente.")
print("Programa terminado")
print("Programa realizado por Mario NL = 53")
```
